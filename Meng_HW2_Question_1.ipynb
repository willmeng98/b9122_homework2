```python
from bs4 import BeautifulSoup
import urllib.request
```

### Problem 1 Part 1

```python
seed_url = "https://press.un.org/en"

urls = [seed_url]    #queue of urls to crawl
seen = [seed_url]    #stack of urls seen so far
opened = []          #we keep track of seen urls so that we don't revisit them
crisis_counter = 0

maxNumUrl = 1000; #set the maximum number of urls to visit
print("Starting with url="+str(urls))
while len(urls) > 0 and len(opened) < maxNumUrl and crisis_counter < 10:
    try:
        curr_url=urls.pop(0)
        req = urllib.request.Request(curr_url,headers={'User-Agent': 'Mozilla/5.0'})
        webpage = urllib.request.urlopen(req).read()
        opened.append(curr_url)
    except Exception as ex:
        continue   
    soup = BeautifulSoup(webpage)
    for tag in soup.find_all('a', href = True):
        # Check for the tag we are looking for
        if "Press Release" in tag.text and "/en/press-release" in tag["href"]:
            print("FOUND", tag)
            # Check if Crisis exist in the content
            page_text = soup.get_text()
            if "crisis" in page_text:
                print("CRISIS")
                print(curr_url)
                crisis_counter += 1
                page_source = webpage.decode('utf-8') 
                with open("1_" + str(crisis_counter)+ ".txt", "w") as text_file:
                    text_file.write(page_source)
                
        childUrl = tag['href'] 
        o_childurl = childUrl
        childUrl = urllib.parse.urljoin(seed_url, childUrl)
        if seed_url in childUrl and childUrl not in seen:
            urls.append(childUrl)
            seen.append(childUrl)
        else:
            pass

print("num. of URLs seen = %d, and scanned = %d" % (len(seen), len(opened)))
```

### Problem 1 Part 2

```python
seed_url = "https://www.europarl.europa.eu/news/en/press-room/page/"

urls = [seed_url]    #queue of urls to crawl
seen = [seed_url]    #stack of urls seen so far
opened = []          #we keep track of seen urls so that we don't revisit them
plenary_counter = 0
page_num = 1

maxNumUrl = 1000; #set the maximum number of urls to visit
print("Starting with url="+str(urls))
while len(opened) < maxNumUrl and plenary_counter < 10 and page_num < 1000:
    page_url = seed_url + str(page_num)
    print("WE ARE ON PAGE ", page_url)
    try:
        req = urllib.request.Request(page_url,headers={'User-Agent': 'Mozilla/5.0'})
        webpage = urllib.request.urlopen(req).read()
        opened.append(page_url)
    except Exception as ex:
        continue   
    soup = BeautifulSoup(webpage)
    current_page_urls = []
    for tag in soup.find_all('a', href = True): 
        childUrl = tag['href'] 
        o_childurl = childUrl
        childUrl = urllib.parse.urljoin(seed_url, childUrl)
        if childUrl not in seen:
            urls.append(childUrl)
            seen.append(childUrl)
            current_page_urls.append(childUrl)
        else:
            pass
    for current_url in current_page_urls:
        try:
            req = urllib.request.Request(current_url,headers={'User-Agent': 'Mozilla/5.0'})
            webpage = urllib.request.urlopen(req).read()
            opened.append(page_url)
        except Exception as ex:
            continue   
        soup = BeautifulSoup(webpage)
        for tag in soup.find_all('span'): #find tags with links
            # Check for the tag we are looking for
            if "Plenary session" in tag.text:
                print("FOUND PLEN", tag, "in", current_url)
                page_text = soup.get_text()
                if "crisis" in page_text:
                    print("CRISIS")
                    print(current_url)
                    plenary_counter += 1
                    page_source = webpage.decode('utf-8') 
                    with open("2_" + str(plenary_counter)+ ".txt", "w") as text_file:
                            text_file.write(page_source)
                    break
        if plenary_counter >= 10:
            break
    page_num += 1

print("num. of URLs seen = %d, and scanned = %d" % (len(seen), len(opened)))
```
